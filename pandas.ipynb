```python
data = {'city': ['Bangalore', 'Bangalore', 'Bangalore',
 'Mumbai', 'Mumbai', 'Mumbai'],
 'year': [2020, 2021, 2022, 2020, 2021, 2022,],
 'population': [10.0, 10.1, 10.2, 5.2, 5.3, 5.5]}
df1 = pd.DataFrame(data)
df1.head()
```
