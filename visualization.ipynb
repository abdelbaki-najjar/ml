#  Introduction to  Data Visualization


```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# The routine np.empty() creates an empty array of given size.
x = np.empty([3, 3], np.uint8)
print(x)
```

```python
# np.eye() to create a matrix of all zeros except the diagonal elements of all the zeros. The diagonal has all the ones.
y = np.eye(4, dtype=np.uint8)
print(y)
```

```python
# We can also set the position of the diagonal as follows
y = np.eye(4, dtype=np.uint8, k=1)
print(y)
```

```python
# np.identity() returns an identity matrix of the specified size. 
x = np.identity(3, dtype= np.uint8)
print(x)
x = np.identity(4, dtype= np.uint8)
print(x)
```

```python
# np.ones() returns the matrix of the given size that has all the elements as ones.
x = np.ones((3, 3, 3), dtype=np.int16)
print(x)
x = np.ones((1, 1, 1), dtype=np.int16)
print(x)
```

```python
# the routine arange(). It creates a Ndarray of evenly spaced values with the given interval. An argument for the stop value is compulsory. The start value and interval parameters have default arguments 0 and 1, respectively.
np.arange(10)
```

```python
# linspace() returns a Ndarray of evenly spaced numbers over a specified interval. We must pass it the starting value, the end value, and the number of values as follows

np.linspace(0, 20, 30)
```

```python
np.logspace(0.1, 2, 10)
```

```python
np.geomspace(0.1, 20, 10)
```

```python
x = np.arange(10)
y = x + 1
plt.plot(x, y)
plt.show()
```

```python
x = np.arange(10)
y1 = 1 - x
plt.plot(x, y, x, y1)
plt.show()
```

```python
n = 3
t = np.arange(0, np.pi*2, 0.05)
y = np.sin( n * t )
plt.plot(t, y)
plt.show()
```

```python
n = 5
x = np.arange(n)
y = np.random.rand(n)
print(y)
plt.bar(x, y)
plt.show()
```

```python
fig, ax = plt.subplots()
ax.bar(x, y)
ax.set_title('Bar Graph')
ax.set_xlabel('X')
ax.set_ylabel('Y')
plt.show()
```

```python
x = np.arange(10)
plt.subplot(3, 3, 1)
plt.plot(x, x)
plt.title('Linear')
plt.subplot(2, 2, 2)
plt.plot(x, x*x)
plt.title('Quadratic')
plt.subplot(2, 2, 3)
plt.plot(x, np.sqrt(x))
plt.title('Square root')
plt.subplot(2, 2, 4)
plt.plot(x, np.log(x))
plt.title('Log')
plt.tight_layout()
plt.show()
```

```python
n = 100
x = np.random.rand(n)
y = np.random.rand(n)
plt.scatter(x, y)
plt.show()
```

```python
 mu, sigma = 0, 0.1
 x = np.random.normal(mu, sigma, 1000)
 plt.hist(x)
 plt.show()
```

```python
 x = np.array([10, 20, 30, 40])
 plt.pie(x)
 plt.show()
```
